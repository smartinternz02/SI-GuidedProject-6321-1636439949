```python
#putting the median in age and mode in cabin column
train.Age = train.Age.fillna(train.Age.median())
train.Cabin = train.Cabin.fillna(train.Cabin.mode())
#train.Embarked = train.Embarked.fillna(train.Embarked.mode())
test.Age = test.Age.fillna(test.Age.median())
test.Cabin = test.Cabin.fillna(test.Cabin.mode())
#test.Fare = test.Fare.fillna(test.Fare.mean())
```

# Label encoding


Sex and embarked has categorical values

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['Sex'] = le.fit_transform(train['Sex'])
train['Embarked'] = le.fit_transform(train['Embarked'])
test['Sex'] = le.fit_transform(test['Sex'])
test['Embarked'] = le.fit_transform(test['Embarked'])
```

```python
train.drop(['Name','Cabin','Ticket'],axis='columns', inplace=True)
test.drop(['Name','Cabin','Ticket'],axis='columns', inplace=True)
```
